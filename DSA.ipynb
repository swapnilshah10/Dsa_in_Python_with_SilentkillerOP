```python
# remove function of bst
from os import *
from sys import *
from collections import *
from math import *

'''
Following is the TreeNode class structure.
'''
class TreeNode:

	def __init__(self, val):
		self.val = val
		self.right = None
		self.left = None

def find_min(node):
    while node.left is not None:
        node = node.left
    return node

	
def remove(root, val):
    if root is None:
        return root

    if val < root.val:
        root.left = remove(root.left, val)
    elif val > root.val:
        root.right = remove(root.right, val)
    else:
        if root.left is None:
            return root.right
        elif root.right is None:
            return root.left
        temp = find_min(root.right)
        root.val = temp.val
        root.right = remove(root.right, temp.val)
    return root


def removeNode(root, X):
	return remove(root , X)
```

```python
# first , last occurance of element in sorted array
def binary_search_first_occurrence(arr, target):
    left, right = 0, len(arr) - 1
    first_occurrence = -1

    while left <= right:
        mid = left + (right - left) // 2

        if arr[mid] == target:
            first_occurrence = mid
            right = mid - 1  # Continue searching on the left side
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1

    return first_occurrence

def binary_search_last_occurrence(arr, target):
    left, right = 0, len(arr) - 1
    last_occurrence = -1

    while left <= right:
        mid = left + (right - left) // 2

        if arr[mid] == target:
            last_occurrence = mid
            left = mid + 1  # Continue searching on the right side
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1

    return last_occurrence

# Example usage
sorted_arr = [1, 2, 2, 2, 3, 4, 4, 5, 5, 5, 6, 7, 8, 8, 9]
target = 5

first_occurrence = binary_search_first_occurrence(sorted_arr, target)
last_occurrence = binary_search_last_occurrence(sorted_arr, target)

print(f"The first occurrence of {target} is at index {first_occurrence}")
print(f"The last occurrence of {target} is at index {last_occurrence}")
```

```python
def print_list(head):
    current = head
    while current:
        print(current.val, end=" -> ")
        current = current.next
    print("None")
# merge two sorted linked list

class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def merge_sorted_lists(l1, l2):
    # Create a dummy node to serve as the head of the merged list
    dummy = ListNode()
    current = dummy

    while l1 and l2:
        if l1.val < l2.val:
            current.next = l1
            l1 = l1.next
        else:
            current.next = l2
            l2 = l2.next
        current = current.next

    # If one list is exhausted, append the remaining nodes of the other list
    if l1:
        current.next = l1
    else:
        current.next = l2

    return dummy.next

# Example usage
# Creating two sorted linked lists
list1 = ListNode(1, ListNode(3, ListNode(5)))
list2 = ListNode(2, ListNode(4, ListNode(6)))

# Merging the lists
print("The merged list is:")

print_list(list1)
print()
print_list(list2)
print()
print_list(merge_sorted_lists(list1, list2))


```

```python
# prefix xor array given find original array
class Solution:
    def findArray(self, pref: List[int]) -> List[int]:
        if len(pref) <= 1: return pref

        t = pref[0]
        for i in range(1 , len(pref)):
            x = pref[i]
            pref[i] = t^pref[i]
            t = x
        return pref
```

```python
class Player:
    def __init__(self, name):
        self.name = name
        self.cards = []

    def add_card(self, card_str):
        self.cards.append(Card(card_str[0], card_str[1:]))

class Card:
    all_values = "2345678910jqka"

    def __init__(self, suit, value):
        self.suit = suit
        self.value = value

    def compare_to(self, other):
        return self.all_values.index(self.value) - self.all_values.index(other.value)

class Game:
    def __init__(self, player_names, trump):
        self.players = {name: Player(name) for name in player_names}
        self.played_cards = []
        self.trump = trump
        self.player_names = player_names

    def add_card_to_player(self, player_name, card_str):
        card = Card(card_str[0], card_str[1:])
        self.players[player_name].add_card(card_str)
        self.played_cards.append(card)

    def get_optimal_card(self, self_player):
        teammate = self.player_names[(self.player_names.index(self_player.name) + 2) % len(self.player_names)]

        highest_card = self.played_cards[0]
        for card in self.played_cards:
            if (highest_card.suit == card.suit and card.compare_to(highest_card) > 0) or (card.suit == self.trump and highest_card.suit != self.trump):
                highest_card = card

        if highest_card in self.players[teammate].cards:
            return self.get_smallest_card(self_player.cards, '0')

        if highest_card.suit != self.trump:
            smallest_bigger_card = self.get_smallest_card_bigger_than(self_player.cards, highest_card, highest_card.suit)
            if smallest_bigger_card is not None:
                return smallest_bigger_card
            return self.get_smallest_card(self_player.cards, '0')

        highest_teammate_trump_card = max((card for card in self.players[teammate].cards if card.suit == self.trump), default=None)
        highest_enemy_trump_card = None

        for player_name in self.player_names:
            if player_name in (teammate, self_player.name):
                continue
            temp_card = max((card for card in self.players[player_name].cards if card.suit == self.trump), default=None)
            if temp_card is not None and (highest_enemy_trump_card is None or temp_card.compare_to(highest_enemy_trump_card) > 0):
                highest_enemy_trump_card = temp_card

        if highest_teammate_trump_card is None or (highest_enemy_trump_card is not None and highest_enemy_trump_card.compare_to(highest_teammate_trump_card) > 0):
            smallest_trump_higher_than_enemy = self.get_smallest_card_bigger_than(self_player.cards, highest_enemy_trump_card, self.trump)
            if smallest_trump_higher_than_enemy is not None:
                return smallest_trump_higher_than_enemy

        return self.get_smallest_card(self_player.cards, '0')

    @staticmethod
    def get_highest_card(cards, suit):
        cards.sort(key=lambda card: Card.all_values.index(card.value), reverse=True)
        for card in cards:
            if card.suit == suit:
                return card
        return None

    @staticmethod
    def get_smallest_card(cards, suit):
        cards.sort(key=lambda card: Card.all_values.index(card.value))
        for card in cards:
            if suit == '0' or card.suit == suit:
                return card
        return None

    @staticmethod
    def get_smallest_card_bigger_than(cards, target_card, target_suit):
        smallest_bigger_card = None
        for card in cards:
            if card.suit == target_suit and card.compare_to(target_card) > 0:
                if smallest_bigger_card is None or card.compare_to(smallest_bigger_card) < 0:
                    smallest_bigger_card = card
        return smallest_bigger_card


if __name__ == "__main__":
    input_str = input().strip()
    input_map = {}
    parts = input_str.split(";")
    for part in parts:
        key, value = part.split(":")
        input_map[key] = value

    player_names = input_map["players"].split(",")
    trump = input_map["trump"]

    game = Game(player_names, trump)

    for player_name in player_names:
        if player_name in input_map:
            game.add_card_to_player(player_name, input_map[player_name])

    self_player = game.players[input_map["self"]]
    self_player.cards = [Card(card_str[0], card_str[1:]) for card_str in input_map["cards"].split(",")]

    optimal_card = game.get_optimal_card(self_player)
    print(optimal_card.suit + optimal_card.value)
```

```python
# Definition for a binary tree node.


# averageOfSubtree
# https://leetcode.com/problems/count-nodes-equal-to-average-of-subtree/submissions/

class Solution:
    def averageOfSubtree(self, root: Optional[TreeNode]) -> int:

        self.ans = 0

        def helper(root):
            if root == None :
                return (0 ,0)

            l = helper(root.left)
            r = helper(root.right)
            if (l[0] + r[0] + root.val) // (l[1] + r[1] + 1) == root.val:
                self.ans+=1 

            return (l[0] + r[0] + root.val , l[1] + r[1] + 1)
        
        helper(root)
        return self.ans
```

```python
# smallest substring containing all characters of another string
def smallest_substring(S, X):
    n = len(S)
    m = len(X)
    if m > n:
        return ""
    freq_X = [0] * 256
    freq_S = [0] * 256
    for i in range(m):
        freq_X[ord(X[i])] += 1
    count = 0
    start = 0
    min_len = n + 1
    min_start = 0
    for j in range(n):
        freq_S[ord(S[j])] += 1
        if freq_X[ord(S[j])] != 0 and freq_S[ord(S[j])] <= freq_X[ord(S[j])]:
            count += 1
        if count == m:
            while freq_S[ord(S[start])] == 0 or freq_S[ord(S[start])] > freq_X[ord(S[start])]:
                if freq_S[ord(S[start])] > freq_X[ord(S[start])]:
                    freq_S[ord(S[start])] -= 1
                start += 1
            len_window = j - start + 1
            if min_len > len_window:
                min_len = len_window
                min_start = start
    if min_len == n + 1:
        return ""
    return S[min_start:min_start + min_len]
```
